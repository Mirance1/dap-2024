4. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

```python
titanic = pd.read_csv('titanic.csv')
pclass_distribution = titanic['Pclass'].value_counts()
print("Распределение по классам для всех пассажиров:")
print(pclass_distribution)

pclass_men = titanic[titanic['Sex'] == 'male']['Pclass'].value_counts()
print("\nРаспределение по классам для мужчин:")
print(pclass_men)


pclass_women = titanic[titanic['Sex'] == 'female']['Pclass'].value_counts()
print("\nРаспределение по классам для женщин:")
print(pclass_women)


men_second_class = titanic[(titanic['Sex'] == 'male') & (titanic['Pclass'] == 2)].shape[0]
print(f"\nКоличество мужчин 2-го класса: {men_second_class}")
```

5. Каковы медиана и среднеквадрати́чное отклонение платежей (Fare)? Округлите до 2 десятичных знаков.

```python

titanic = pd.read_csv('titanic.csv')

median_fare = round(titanic['Fare'].median(), 2)
std_fare = round(titanic['Fare'].std(), 2)

print(f"Медиана платежей (Fare): {median_fare}")
print(f"Среднеквадратичное отклонение платежей (Fare): {std_fare}")
```

6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
titanic = pd.read_csv('titanic.csv')

group_under_30 = titanic[titanic['Age'] < 30]
group_over_60 = titanic[titanic['Age'] > 60]

survival_under_30 = group_under_30['Survived'].mean()
survival_over_60 = group_over_60['Survived'].mean()

print(f"Доля выживших среди пассажиров младше 30 лет: {survival_under_30:.2f}")
print(f"Доля выживших среди пассажиров старше 60 лет: {survival_over_60:.2f}")
```

7. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python

titanic = pd.read_csv('titanic.csv')

group_men = titanic[titanic['Sex'] == 'male']
group_women = titanic[titanic['Sex'] == 'female']


survival_men = group_men['Survived'].mean()
survival_women = group_women['Survived'].mean()


print(f"Доля выживших среди мужчин: {survival_men:.2f}")
print(f"Доля выживших среди женщин: {survival_women:.2f}")
```

8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

10. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
titanic = pd.read_csv('titanic.csv')

titanic = titanic.dropna(subset=['Age'])

mean_age_survived = titanic[titanic['Survived'] == 1]['Age'].mean()
mean_age_not_survived = titanic[titanic['Survived'] == 0]['Age'].mean()

print(f"Средний возраст спасённых пассажиров: {mean_age_survived:.2f}")
print(f"Средний возраст погибших пассажиров: {mean_age_not_survived:.2f}")
print(f"Средний возраст погибших выше спасённых? {mean_age_not_survived > mean_age_survived}")
```
